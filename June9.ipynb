3. For fields with many missing values, determine which ones have enough useful values to
keep and which ones you should avoid using in your analysis.

We first calculate the number of missing values and the percentage of missing values in each column. We can assume that the columns with less than 50% of missing values would be useful in our analysis.
So we discard the rest of the columns(with more than 50% missing values) from our analysis.

We write a function null_count to apply for each column from the Immidata dataframe.

```python
def null_count(column):
    # Make a vector that contains True if null, False if not.
    column_null = pd.isnull(column)
    # Create a new vector with only values where the series is null.
    null = column[column_null == True]
    # Return the percentage of null values.
    return ((len(null) * 100) / (ImmiData.shape[0] - 1))


column_null_count = ImmiData.apply(null_count)
print(column_null_count)
```
